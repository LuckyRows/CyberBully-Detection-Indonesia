```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = pd.read_csv('dataset/dataset_komentar_instagram_cyberbullying.csv')
dataset
```

```python
dt_transformed = dataset[['Sentiment', 'Instagram Comment Text']]
y = dt_transformed.iloc[:, :-1].values
y
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
y = np.array(ct.fit_transform(y))
```

```python
print(y)
```

```python
y_df = pd.DataFrame(y)
y_hate = np.array(y_df[0])
y_offensive = np.array(y_df[1])
```

```python
print(y_hate)
print(y_offensive)
```

```python
corpus = []
for i in range (0, 400):
    review = re.sub("[^a-zA-Z]", ' ', dt_transformed['Instagram Comment Text'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('indonesian')
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
```

```python
cv = CountVectorizer(max_features = 2000)
x = cv.fit_transform(corpus).toarray()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_hate, test_size = 0.2, random_state = 0)
```

```python
classifier_rf = RandomForestClassifier()
classifier_rf.fit(x_train, y_train)
```

```python
classifier_np = GaussianNB()
classifier_np.fit(x_train, y_train)
```

```python
classifier_lr = LogisticRegression(random_state = 0)
classifier_lr.fit(x_train, y_train)
```

```python
classifier_svm = svm.SVC()
classifier_svm.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred_rf = classifier_rf.predict(x_test)
cm = confusion_matrix(y_test, y_pred_rf)
print(cm)
```

```python
y_pred_np = classifier_np.predict(x_test)
cm = confusion_matrix(y_test, y_pred_np)
print(cm)
```

```python
y_pred_svm = classifier_svm.predict(x_test)
cm = confusion_matrix(y_test, y_pred_svm)
print(cm)
```

```python
y_pred_lr = classifier_lr.predict(x_test)
cm = confusion_matrix(y_test, y_pred_lr)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
rf_score = accuracy_score(y_test, y_pred_rf)
svm_score = accuracy_score(y_test, y_pred_svm)
lr_score = accuracy_score(y_test, y_pred_lr)
np_score = accuracy_score(y_test, y_pred_np)

print('Random Forest Accuracy:', str(rf_score))
print('Support Vector Machine Accuracy:', str(svm_score))
print('Logistic Regression Accuracy:', str(lr_score))
print('Naive Bayes Accuracy:', str(np_score))
```
